```python
import random
arr = ["r","p","s"]
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

```

```python
p1_score=0
p2_score=0
```

```python
while(True):
    p1=input("WHAT WOULD YOU LIKE TO CHOOSE? Press r for ROCK,p for PAPER and s for Scissors (q to exit)")
    if p1 not in arr:
        print("choose correctly")
    if(p1=='q'):
        break
    p2 = random.choice(arr)    
    if (p1 =='r' and p2 =="s")  or (p1=='p' and p2 =='r') or (p1=='s' and p2 =='p') :
        print("player scored")
        p1_score+=1
    elif (p1=='r' and p2 =='s') or (p1=='p' and p2 =='s') or (p1=='s' and p2 =='r'):
        print("computer scored")
        p2_score+=1
    else:
        print("draw")
        p1_score+=1
        p2_score+=1
    print("player chose : ",p1)
    if(p1=='r'):
        print(rock)
    elif(p1=='p'):
        print(paper)
    elif(p1=='s'):
        print(scissors)
    print("computer chose : ",p2 )
    if(p2=='r'):
        print(rock)
    elif(p2=='p'):
        print(paper)
    elif(p2=='s'):
        print(scissors)
    print("player score : ",p1_score , "| computer score : ",p2_score )

```
