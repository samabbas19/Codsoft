```python
import tkinter
from tkinter import *

root = Tk()
root.title("TODO APP")
root.geometry("400x650+400+100")
root.resizable(False,False)

#task
task_list = []

def deletetask():
    global task_list
    selected_task = list1.get(ANCHOR)
    if selected_task:
        task_list.remove(selected_task)
        list1.delete(ANCHOR)

        with open(r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\tasklist.txt", "w") as taskfile:
            for task in task_list:
                taskfile.write(f"{task}\n")

def addtask():
    task = taskentry.get()
    taskentry.delete(0,END)
    if task:
        taskfile = open(r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\tasklist.txt","a")
        taskfile.write(f"\n{task}")
        task_list.append(task)
        list1.insert(END,task)

def opentask():
    taskfile = open(r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\tasklist.txt","r")
    tasks = taskfile.readlines()
    
    for t in tasks:
        if t!="\n":
            task_list.append(t)
            list1.insert(END,t)
    taskfile.close()

#icon
icon = PhotoImage(file=r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\Image\task.png")
header = PhotoImage(file=r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\Image\topbar.png")
dock  = PhotoImage(file=r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\Image\dock.png")
note = PhotoImage(file=r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\Image\task.png")
delete = PhotoImage(file=r"C:\Users\saame\OneDrive\Desktop\SEM 3\PAI\PRACTICE\ANGELINA BOOTCAMP\Tkinter\Image\delete.png")


root.iconphoto(False,icon)
Label(root,image=header).pack()
Label(root,image=dock,bg="#32405b").place(x=30,y=25)
# Label(root,image=note,bg="#32405b").place(x=30,y=25)

heading = Label(root,text="TO-DO-LIST APP",font="aerial 20 bold",fg="white",bg="#32405b").place(x=100,y=18)

#working
frame = Frame(root,width=400,height=50,bg="white")
frame.place(x=0,y=100)

task = StringVar()
taskentry = Entry(frame,width=18,font = "aerial 20",bd=0)
taskentry.place(x=10,y=7)
taskentry.focus()

#button
button = Button(frame,text="ADD",font="aerial 20 bold",width =6,bg="#5a95ff",fg="#fff",bd=0,command=addtask)
button.place(x=300,y=0)

# #list
frame1 = Frame(root,bd=3,width=700,height=200,bg="#32405b")
frame1.pack(pady=(160,0))

list1 = Listbox(frame1,font="aerial 20 ",width=40,height=10,bg="#32405b",fg="white",cursor="hand2",selectbackground="#5a95ff")
list1.pack(side=LEFT,fill=BOTH,padx=2)

scroll = Scrollbar(frame1)
scroll.pack(side=RIGHT,fill=BOTH,)

list1.config(yscrollcommand=scroll.set)
scroll.config(command=list1.yview)

opentask()

# delete button
delete_button = Button(root, image=delete, bd=0,command=deletetask)
delete_button.pack(side=BOTTOM, pady=13)
delete_button.lift()



root.mainloop()
```
